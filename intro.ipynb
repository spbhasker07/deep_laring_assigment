```python
import tensorflow as tf
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
X = iris.data     
y = iris.target
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
ol=Dense(units=3,activation='softmax')
```

```python
model.add(ol)
```

```python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
model.fit(X,y,batch_size=42,epochs=200)
```

```python
import numpy as np
sample = np.array([5.1, 3.5, 1.4, 0.2])  # example iris flower
prediction = model.predict(sample.reshape(1, 4))
print(prediction)
print("Predicted class:", np.argmax(prediction))
```
