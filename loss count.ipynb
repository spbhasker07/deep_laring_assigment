```python
from sklearn.datasets import load_iris
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
import tensorflow as tf
import numpy as np
import random
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
tf.random.set_seed(10)
random.seed(10)
np.random.seed(10)
```

```python
model=Sequential()
```

```python
model.add(Dense(units=3,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer=tf.keras.optimizers.SGD(learning_rate=.1),metrics=['accuracy'])
```

```python
model.fit(x,y,batch_size=100,epochs=200)
```

```python
losses=model.history.history['loss']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,201),losses)
plt.xlabel('epoch')
plt.ylabel('losses')
plt.show()
```
