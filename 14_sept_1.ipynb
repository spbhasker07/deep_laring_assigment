```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import random
```

```python
df=pd.read_excel('creditcard_fraud.xlsx')
df
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X)
```

```python
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

model=Sequential()
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=56,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','precision','recall'])
model.fit(X_train,y,batch_size=32,epochs=10)
```

```python
p=np.array(model.history.history['precision'])
r=np.array(model.history.history['recall'])
f1=2*p*r/(p+r)
print(f1)
```

```python
df=pd.read_csv('f:/ML_Dataset/ncr_house_price.csv')
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.city.value_counts()
```

```python
df.city=df.city.map({'Bengaluru':0,'Gurugram':1,'Noida':2})
df
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X)
```

```python
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

model=Sequential()

model.add(Dense(units=512,activation='relu'))
model.add(Dense(units=256,activation='relu'))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=1,activation='linear'))  #no activation

model.compile(loss='mse',optimizer='adam',metrics=['r2_score'])
model.fit(X_train,y,batch_size=32,epochs=20,validation_split=.2)
```

```python
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

model=Sequential()

model.add(Dense(units=512,activation='relu'))
model.add(Dense(units=256,activation='relu'))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=1,activation='linear'))  #no activation

model.compile(loss='mae',optimizer='adam',metrics=['r2_score'])
model.fit(X_train,y,batch_size=32,epochs=20,validation_split=.2)
```

```python
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

model=Sequential()

model.add(Dense(units=512,activation='relu'))
model.add(Dense(units=256,activation='relu'))
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=1,activation='linear'))  #no activation

model.compile(loss='huber',optimizer='adam',metrics=['r2_score'])
model.fit(X_train,y,batch_size=32,epochs=20,validation_split=.2)
```
