```python
def create_data(data, input_steps, output_steps):
    inputs = []
    targets = []
    for i in range(len(data) - input_steps - output_steps + 1):
        inputs.append(data[i : i + input_steps]) # Input window
        targets.append(data[i + input_steps : i + input_steps + output_steps]) # O
    return np.array(inputs), np.array(targets)
```
